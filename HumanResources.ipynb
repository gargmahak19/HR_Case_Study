# Create Train and Test Data and Data Cleaning 

```python
employee_df.head(3)
```

##### We have to first convert the catagorical data into number or else the model with not take them as an input.

```python
x_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender','JobRole', 'MaritalStatus']]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
x_cat = encoder.fit_transform(x_cat).toarray()
```

```python
x_cat = pd.DataFrame(x_cat)
```

```python
x_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]

```

```python
x_all = pd.concat([x_cat, x_numerical], axis = 1)

x_all.columns = x_all.columns.astype(str)
```

##### We have encoded the whole data into but there were already data with numbers present so we need to do something about it as the scaling is different. 

##### We have data in binary and we have data like employee's incomes. So we need to scale the entire table for the model to treat every data the same way.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scale = scaler.fit_transform(x_all)
```

```python
x = pd.DataFrame(x_scale)
```

```python
y = employee_df['Attrition']
```

# Model Training

## Logistic Regression 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)

y_predict = model.predict(x_test)
y_predict = pd.DataFrame(y_predict)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy {} %".format( 100 * accuracy_score(y_predict, y_test)))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_predict, y_test)
sns.heatmap(cm, annot = True)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_predict, y_test))
```

## Deep Learning Model

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
# Here keras is the API for tensorflow which will do all the backend work for tensorflow. 
# Sequential is like how we see layers after layers getting created.

model.add(tf.keras.layers.Dense(units = 500, activation = 'relu', input_shape = (50,)))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# For the last layer we keep only one unit and the activation function will be sigmoid.
```

```python
# Here we can see how 4 layers are created and each layer has weights and how many weights it has.
```

```python
model.summary()
```

```python
model.compile(optimizer = "Adam", loss = "binary_crossentropy", metrics = ['accuracy'])
```

```python
eochs_hist = model.fit(x_train, y_train, epochs = 100, batch_size = 50)
```

```python
y_pred = model.predict(x_test)
```

```python
eochs_hist.history.keys()
```

```python
plt.plot(eochs_hist.history['loss'])
plt.title("Model Loss progress during Training")
plt.xlabel = ("Epochs")
plt.ylabel = ("Training Loss")
plt.legend(['Training Loss'])
```

```python
plt.plot(eochs_hist.history['accuracy'])
plt.title("Model Accuracy progress during Training")
plt.xlabel = ("Epochs")
plt.ylabel = ("Training Accuracy")
plt.legend(['Training Accuracy'])
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred.round(), normalize = 'all')
sns.heatmap(cm, annot = True)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_pred.round(), y_test))
```
